```python
total_revenue = df['revenue'].sum()
total_revenue
```

```python
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M')

monthly_revenue = df.groupby('order_month')['revenue'].sum()
monthly_revenue
```

```python
monthly_revenue.plot(kind='line', figsize=(10,5), title='Monthly Revenue Trend')
```

```python
cancelled = df[df['order_status'] == 'canceled']
cancelled_loss = cancelled['revenue'].sum()

cancelled_loss
```

```python
df['late_delivery'] = (
    df['order_delivered_customer_date'] >
    df['order_estimated_delivery_date']
)
df['late_delivery'].value_counts()
```

```python
customer_orders = df.groupby('customer_id')['order_id'].nunique()
```

```python
customer_type = customer_orders.apply(
    lambda x: 'Repeat' if x > 1 else 'One-time'
)
```

```python
df = df.merge(customer_type.rename('customer_type'),
              left_on='customer_id',
              right_index=True)

df.groupby('customer_type')['revenue'].sum()
```
